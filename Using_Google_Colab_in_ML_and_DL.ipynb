# Extensões úteis do Colab

O Colab também vem com muitas extensões realmente úteis. Uma dessas extensões nos permite montar nosso Google Drive em nosso diretório de trabalho. Podemos fazer isso usando:

```python
import os
from google.colab import drive

 
MOUNTPOINT = "/content/gdrive"
DATADIR = os.path.join(MOUNTPOINT, "MyDrive")
drive.mount(MOUNTPOINT)
```

Em seguida, o Colab solicitará permissão para acessar seus arquivos do Google Drive, o que você pode fazer após selecionar a qual conta do Google deseja conceder acesso. Depois de dar as permissões necessárias, podemos ver nosso Google Drive montado na guia Arquivos no lado esquerdo.

image.png

```python
# writes directly to google drive
with open(f"{DATADIR}/test.txt", "w") as outfile:
    outfile.write("Hello World!")
```

Este ``snippet`` de código grava ``Hello World!`` em um arquivo ``test.txt`` no nível superior do seu Google Drive. Da mesma forma, também podemos ler um arquivo em nosso Google Drive usando:

```python

with open(f"{DATADIR}/test.txt", "r") as infile:
    file_data = infile.read()
    print(file_data)
```

Além disso, o Google Colab vem com algumas extensões para criar uma melhor experiência usando um notebook. Se usarmos muito o ``pandas DataFrame``, há uma extensão para exibir tabelas interativas. Para usar isso, podemos usar funções mágicas:

```python
%load_ext google.colab.data_table
```

Isso habilita a exibição interativa para ``DataFrames``, então quando executamos:

```python
from sklearn.datasets import fetch_openml

X = fetch_openml("diabetes", version=1, as_frame=True, return_X_y=False)["frame"]
X
```

```python
X.shape
```

Isso mostrará o ``DataFrame`` como uma tabela interativa, onde ``podemos filtrar com base em colunas``, ver as diferentes linhas da tabela, etc.


Para desativar esse recurso mais tarde, podemos executar:

```python
%unload_ext google.colab.data_table
```

Quando exibimos o mesmo ``DataFrame X``, novamente, obtemos a interface padrão do Pandas DataFrame:

```python
X
```
